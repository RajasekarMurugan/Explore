```python
import nltk
from nltk.corpus import movie_reviews # Pre-built dataset
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.classify import NaiveBayesClassifier #  Classfification Algorithm
from nltk.classify.util import accuracy
import random
```

```python
nltk.download("movie_reviews")
nltk.download("stopwords")
nltk.download("punkt")
nltk.download("punkt_tab")
```

```python
# Stopwords removal (Keep only the keywords)
def extract_features(words):
  stop_words = set(stopwords.words("english")) # setting the stopwords to english
  words = [word for word in words if word.lower() not in stop_words]
  return {word : True for word in words}
```

```python
reviews = [(list(movie_reviews.words(fileid)),category)
for category in movie_reviews.categories()
for fileid in movie_reviews.fileids(category)]
```

```python
random.shuffle(reviews)
```

```python
feature_sets = [(extract_features(review), category) for review, category in reviews]
```

```python
train_size = int(len(feature_sets)*0.8)
train_set, test_set = feature_sets[:train_size],feature_sets[train_size:]
```

```python
classifier = NaiveBayesClassifier.train(train_set)
```

```python
classifier.show_most_informative_features(10)
```

```python
def predict_sentiment(review):
  words = word_tokenize(review)
  features = extract_features(words)
  return classifier.classify(features)
```

```python
test_review = "the move was absolutley amazing, great performance and plot"
```

```python
predict_sentiment(test_review)
```
